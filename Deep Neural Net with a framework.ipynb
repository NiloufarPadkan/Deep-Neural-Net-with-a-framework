### ***Preparing data***

```python
import random
from keras.constraints import maxnorm
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import metrics
from keras.utils.np_utils import to_categorical
import tensorflow as tf
from keras.models import Sequential
from sklearn.metrics import confusion_matrix
from keras.layers import Dense, Dropout, BatchNormalization
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import classification_report
from sklearn.metrics import recall_score, precision_score, f1_score
from sklearn.metrics import precision_recall_fscore_support, accuracy_score
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split

!gdown --id 1-Zyp-JP3f9QhPKaErBkPPFNKaPS1v74u

categorical_attr = ['gender', 'NationalITy', 'PlaceofBirth', 'StageID', 'GradeID', 'SectionID', 'Topic', 'Semester', 'Relation', 'ParentAnsweringSurvey', 'ParentschoolSatisfaction', 'StudentAbsenceDays', 'Class']
```

```python
df = pd.read_csv('/content/Dataset.csv')
df.head()
```

```python
print(df.shape)
```

```python
# Converting Categorical values to scaler values
le = LabelEncoder()
df[categorical_attr] = df[categorical_attr].apply(le.fit_transform, axis=0)
```

```python
# X: Features, y: Classes
X = np.array(df.iloc[:, :-1])
y = np.array(df['Class'])
```

```python
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.2, random_state=12)

print(y_train)
```

```python
y_train = to_categorical(y_train)
y_val = to_categorical(y_val)
print(y_train)
```

```python
print('Number of dataset: ', len(X))
print('Number of train set: ', len(X_train))
print('Number of validation set: ', len(X_val))
```

```python
seed = 7
np.random.seed(seed)
```

## 3 hidden layer

# **Adam optimizer**  

## relu+softmax


```python
tf.keras.backend.clear_session()

optmzr = tf.keras.optimizers.Adam(learning_rate=0.001)

# defining the keras model
AdamReluModel = Sequential()
AdamReluModel.add(Dense(64, input_dim=16, activation='relu'))
AdamReluModel.add(Dense(32, activation='relu'))
AdamReluModel.add(Dense(24, activation='relu'))
AdamReluModel.add(Dense(16, activation='relu'))
AdamReluModel.add(Dense(12, activation='relu'))
AdamReluModel.add(Dense(8, activation='relu'))
AdamReluModel.add(Dense(3, activation='softmax'))

# compile
AdamReluModel.compile(loss='categorical_crossentropy', optimizer=optmzr, metrics=['accuracy'])

# fit the keras model on the dataset
AdamReluHistory = AdamReluModel.fit(X_train, y_train, epochs=45,validation_data=(X_val, y_val))


_, accuracy = AdamReluModel.evaluate(X_train, y_train)
print('Accuracy of train set: %.2f' % (accuracy*100))

_, accuracy = AdamReluModel.evaluate(X_val, y_val)
print('Accuracy of test set: %.2f' % (accuracy*100))



# summarize history for accuracy
plt.plot(AdamReluHistory.history['accuracy'])
plt.plot(AdamReluHistory.history['val_accuracy'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'validation'], loc='upper left')
plt.show()

# summarize history for loss
plt.plot(AdamReluHistory.history['loss'])
plt.plot(AdamReluHistory.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'validation'], loc='upper left')
plt.show()



y_pred = AdamReluModel.predict(X_val)
y_pred = np.argmax(y_pred, axis=1)
y_test = np.argmax(y_val, axis=1)
print(y_pred)


# Print the precision and recall, among other metrics
print(classification_report(y_test, y_pred, digits=3))

# Print the confusion matrix
print(confusion_matrix(y_test, y_pred))



print(AdamReluHistory.history.keys())
AdamReluModel.summary()
```
